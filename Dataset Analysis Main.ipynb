```python

import numpy as np
# Calculating the mean F1 score and mean Imbalance Ratio for each distinct dataset and algorithm
mean_f1_imbalance_by_algo_dataset = scores_data.groupby(['Dataset', 'Algorithm']).agg({'F1': 'mean', 'Imbalance Ratio': 'mean'}).reset_index()

# Enhancing the sorting to ensure datasets are ordered by imbalance ratio then by dataset for consistent x-axis arrangement
mean_f1_imbalance_by_algo_dataset['Sort Key'] = mean_f1_imbalance_by_algo_dataset.groupby(['Dataset'])['Imbalance Ratio'].transform('mean')
mean_f1_imbalance_by_algo_dataset_sorted = mean_f1_imbalance_by_algo_dataset.sort_values(['Sort Key', 'Dataset', 'Algorithm'])

# Finding unique datasets and algorithms for plotting
unique_datasets = mean_f1_imbalance_by_algo_dataset_sorted['Dataset'].unique()
unique_algorithms = mean_f1_imbalance_by_algo_dataset_sorted['Algorithm'].unique()

# Preparing plot data
n_datasets = len(unique_datasets)
n_algorithms = len(unique_algorithms)
bar_width = 0.8 / n_algorithms  # Dynamic width to accommodate multiple bars side by side
index = np.arange(n_datasets)

# Creating the bar plot
plt.figure(figsize=(14, 8))

for i, algorithm in enumerate(unique_algorithms):
    subset = mean_f1_imbalance_by_algo_dataset_sorted[mean_f1_imbalance_by_algo_dataset_sorted['Algorithm'] == algorithm]
    positions = index + (i - n_algorithms / 2) * bar_width + bar_width / 2  # Centering bars around the tick
    plt.bar(positions, subset['F1'], width=bar_width, label=algorithm)

plt.xlabel('Dataset',fontsize='large')
plt.ylabel('Mean F1',fontsize='large')
plt.title('Mean F1 by Dataset and Algorithm (Ordered by Imbalance Ratio)',fontsize='large')
plt.xticks(index, unique_datasets, rotation=45, ha="right",fontsize='large')
plt.gca().tick_params(axis='y', labelsize='large')  # Adjusting the fontsize of the y-axis labels

plt.legend(fontsize='large')
plt.tight_layout()

plt.show()
```
