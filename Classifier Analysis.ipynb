```python
#Load the data 
import pandas as pd
scores_data = pd.read_csv('scores_main_combinations.csv')
scores_data.head()
```

```python
#Dropping the indexed column
column_name_to_remove = 'Unnamed: 0'
scores_data = scores_data.drop(column_name_to_remove, axis=1)
scores_data.head()
scores_data.tail()
```

```python
imbalance_data_dict = {
    'setapProcessT1_updated1.csv': 2.2,
    'setapProcessT2_updated.csv': 1.96,
    'setapProcessT3_updated.csv': 1.96,
    'setapProcessT4_updated.csv': 2.50,
    'setapProcessT5_updated.csv': 1.96,
    'setapProcessT6_updated.csv': 1.96,
    'setapProcessT7_updated.csv': 1.96,
    'setapProcessT8_updated.csv': 1.96,
    'setapProcessT9_updated.csv': 1.96,
    'setapProcessT10_updated.csv': 1.96,
    'setapProcessT11_updated.csv': 1.96,
    'period_data_updated.csv': 2.33,
    'darwin_data_updated.csv': 1.05,
    'toxicity_data_updated.csv': 2.05,
    'voice_data.csv': 2.00,
    'colon 2000.csv': 1.82,
    'DLBCL.csv': 3.05,
    'gastric cancer.csv': 2.75,
    'gastroenterology.csv': 2.62,
    'leukemia.csv': 1.88
}
```

```python
# Assuming you have a DataFrame named df and the imbalance_data_dict as provided
scores_data['Imbalance Ratio'] = scores_data['Dataset'].map(imbalance_data_dict)
scores_data.head()

```

```python
# Updating the 'Dataset' column values to "T1" where the current value is "setapProcessT1_updated1.csv"
scores_data.loc[scores_data['Dataset'] == 'setapProcessT1_updated1.csv', 'Dataset'] = 'T1'
scores_data.loc[scores_data['Dataset'] == 'setapProcessT2_updated.csv', 'Dataset'] = 'T2'
scores_data.loc[scores_data['Dataset'] == 'setapProcessT3_updated.csv', 'Dataset'] = 'T3'
scores_data.loc[scores_data['Dataset'] == 'setapProcessT4_updated.csv', 'Dataset'] = 'T4'
scores_data.loc[scores_data['Dataset'] == 'setapProcessT5_updated.csv', 'Dataset'] = 'T5'
scores_data.loc[scores_data['Dataset'] == 'setapProcessT6_updated.csv', 'Dataset'] = 'T6'
scores_data.loc[scores_data['Dataset'] == 'setapProcessT7_updated.csv', 'Dataset'] = 'T7'
scores_data.loc[scores_data['Dataset'] == 'setapProcessT8_updated.csv', 'Dataset'] = 'T8'
scores_data.loc[scores_data['Dataset'] == 'setapProcessT9_updated.csv', 'Dataset'] = 'T9'
scores_data.loc[scores_data['Dataset'] == 'setapProcessT10_updated.csv', 'Dataset'] = 'T10'
scores_data.loc[scores_data['Dataset'] == 'setapProcessT11_updated.csv', 'Dataset'] = 'T11'
scores_data.loc[scores_data['Dataset'] == 'darwin_data_updated.csv', 'Dataset'] = 'darwin'
scores_data.loc[scores_data['Dataset'] == 'period_data_updated.csv', 'Dataset'] = 'period'
scores_data.loc[scores_data['Dataset'] == 'voice_data_updated.csv', 'Dataset'] = 'voice'
scores_data.loc[scores_data['Dataset'] == 'toxicity_data_updated.csv', 'Dataset'] = 'toxicity'
scores_data.loc[scores_data['Dataset'] == 'colon 2000.csv', 'Dataset'] = 'colon 2000'
scores_data.loc[scores_data['Dataset'] == 'leukemia.csv', 'Dataset'] = 'leukemia'
scores_data.loc[scores_data['Dataset'] == 'gastroenterology.csv', 'Dataset'] = 'gastroenterology'
scores_data.loc[scores_data['Dataset'] == 'gastric cancer.csv', 'Dataset'] = 'gastric cancer'
scores_data.loc[scores_data['Dataset'] == 'DLBCL.csv', 'Dataset'] = 'DLBCL'
scores_data.loc[scores_data['Algorithm'] == 'ReliefFTwo', 'Algorithm'] = 'RiceVD'


# Displaying the updated DataFrame to confirm the change
scores_data
```
